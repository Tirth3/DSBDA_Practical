# Importing required libraries

```python
import pandas as pd
```

# Loading dataset
- If the dataset is in xlsx format, pd.read_excel() and pd.DataFrame() methods are used
- else pd.read_csv() is used
- And then printing columns , shape and head of dataset

```python
df = pd.DataFrame(pd.read_excel("student.xlsx")) 

#df = pd.read_csv("student.csv") 

print(df.columns)
print(df.shape)
df.head
```

# Handling missing values in the dataset
- using fillna() method to fill the NA values with required values(mean , mode , std. deviation etc.)

```python
df.isnull().sum()
df.fillna(0)
```

```python
df.dtypes
```

# converting object datatype(string) to categorical datatype

```python
df['Branch'] = df['Branch'].astype("category")
df['Gender'] = df['Gender'].astype("category")
df['Board[10th]'] = df['Board[10th]'].astype("category")
df['Board[12th]'] = df['Board[12th]'].astype("category")
df['Category'] = df['Category'].astype("category")

cat_cols = df.select_dtypes(['category']).columns
cat_cols
```

# converting categorical values to qualitative values

```python
df[cat_cols] = df[cat_cols].apply(lambda x: x.cat.codes)
df.dtypes
```
