```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame(pd.read_excel('student.xlsx'))
print(df.columns)
df
```

```python
df = df.dropna()
df
```

```python
sns.boxplot(df['GPA 1'])
```

```python
Q1 = df['GPA 1'].quantile(0.25)
Q3 = df['GPA 1'].quantile(0.75)
IQR = Q3 - Q1
print("IQR : " , IQR)
lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR
print("Upper bound: " , upper)
print("Lower bound: ", lower)
```

```python
df_outliers_up = (df['GPA 1'] > upper)
df_outliers_low = (df['GPA 1'] < lower)

out_df = df[~(df_outliers_up | df_outliers_low)]
out_df
```

```python
sns.boxplot(out_df['GPA 1'])
```
