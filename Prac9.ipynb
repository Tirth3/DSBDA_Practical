```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic = sns.load_dataset('titanic')
df = pd.DataFrame(titanic)
df
```

```python
df['age'].fillna(df['age'].mean())
```

```python
# Plot boxplot
sns.boxplot(data=titanic, x='sex', y='age', hue='survived')
plt.title("Age Distribution by Gender and Survival Status")
plt.xlabel("Gender")
plt.ylabel("Age")
# plt.legend(title='Survived', labels=['No', 'Yes'])  # Survival: 0 = No, 1 = Yes
plt.show()
```

# The box shows:

- Median (center line)

- IQR (box edges: 25thâ€“75th percentile)

- Whiskers (range, limited by 1.5x IQR)

- Outliers (points outside whiskers)

# Observations

## 1. Survival by Gender:
- Females had a significantly higher survival rate than males.
- For males, the opposite is true â€” more men died than survived.

## 2. Age Distribution Patterns:
- Children (younger ages) are more likely to appear in the survived group, especially among females.
- Non-survivors (both males and females) have wider age ranges, but for males, many were in the 20â€“40 range.

## 4. Outliers and Spread:
- Several older passengers (age > 60) survived, especially among females â€” these are seen as outliers above the whiskers.
- The spread of ages (IQR) is slightly narrower for survivors than non-survivors, indicating more consistent age patterns among those who survived.
