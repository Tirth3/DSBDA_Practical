```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Iris.csv')
df
```

```python
df.isnull().sum()
```

```python
print("Features : " , df.columns)
```

### sepal length (cm) : Numeric : Length of the sepal in centimeters
### sepal width (cm) : Numeric : Width of the sepal in centimeters
### petal length (cm) : Numeric : Length of the petal in centimeters
### petal width (cm) : Numeric : Width of the petal in centimeters
### target (species) : Nominal : Categorical class (Setosa, Versicolor, Virginica)

```python
numeric_features = df.columns[1:-1]

df[numeric_features].hist(bins=15, figsize=(10, 6), color='pink', edgecolor='black')
plt.suptitle("Histograms of Iris Features")
plt.tight_layout()
plt.show()
```

# Observations from Histograms:

## - Petal length and width show clear separation/clusters, likely linked to species.

## - Sepal width has a somewhat symmetric distribution.

## - Sepal length is slightly right-skewed.

## - No sharp outliers visible in histograms, but further clarity will come with box plots.

```python
plt.figure(figsize=(12, 6))
for i, feature in enumerate(numeric_features):
    plt.subplot(1, 4, i + 1)
    sns.boxplot(y=df[feature], color='lightgreen')
    plt.title(feature)
    plt.tight_layout()
plt.suptitle("Box Plots of Iris Features", y=1.05)
plt.show()
```

# Box Plot Observations:

## - Petal width and length show very distinct groupings and clear IQR differences, indicating high predictive power for classification.

## - Sepal width shows some outliers (low values < 2.2 and high > 4.0).

## - Petal features appear cleanly split, with very few outliers.

## - Sepal length has a few mild outliers on the lower side.
