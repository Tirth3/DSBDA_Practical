# Importing required libraries

```python
import pandas as pd
import numpy as np
```

# Loading the Iris dataset and displaying first 5 rows

```python
df = pd.read_csv('Iris.csv')
df.head()
```

# Calculating statistical summary for every column using df.describe() method

Else one can use different method like df.mean() , df.median() , df.std() etc.

```python
df.describe()
```

# Calculating statistical summary for every column grouped by category i.e specie

```python
statistics_grouped_by_species = df.groupby('Species')['SepalLengthCm'].agg(['mean' , 'median' , 'std' , 'var' , 'min' , 'max' , 'count' , 
                                                                           ('25%' , lambda x : np.percentile(x, 25)), 
                                                                           ('50%' , lambda x : np.percentile(x, 50)),
                                                                           ('75%' , lambda x : np.percentile(x, 75))])
statistics_grouped_by_species
```
